```r
library(tidyverse)
library(repr)
library(tidymodels)
```

```r
crime_data = read_csv("crime.csv") %>%
  filter(YEAR == "2017", TYPE == "Mischief")
crime_data
```

```r
crime_data = as.data.table(crime_data)
crime_data
```

```r
crime_tab = crime_data[NEIGHBOURHOOD!=""]
crime_tab
```

```r
plot(crime_tab, pch = 20, col = alpha("#F9B7DA",0.50))
```
